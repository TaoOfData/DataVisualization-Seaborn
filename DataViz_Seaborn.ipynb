# Flight Delays Data

```python
flight_data = pd.read_csv("flight_delays.csv",index_col="Month")
flight_data
```

## Bar Plot

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
#Cannot use 'Month' as column, instead use the index (Month is Index in dataset)
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

## Heatmap

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

## Insurance Data

```python
# Path of the file to read
insurance_filepath = "insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
#Adding a regression line
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#Adding separate regression lines based on hue
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
#Categorical scatter plot - swarmplot

sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

# Histograms and Density Plots

```python
#Iris data
# Path of the file to read
iris_filepath = "iris.csv"

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

## Histogram

```python
# Histogram 
sns.histplot(iris_data['Petal Length (cm)'])
```

## Kernel Density Estimate Plot (KDE) - Smoothed Histogram

```python
# KDE plot 
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

## 2D KDE plot

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'],hue=iris_data['Species'], kind="kde",shade=True)
```

## Color-coded plots

```python
# Histograms for each species
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
